```python
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import re
stop_words = set(stopwords.words('english'))
other_sw = ['arent', 'amnt', 'isnt', 'youre', 
            'im', 'shes', 'hes', 'were', 'theyre', 
            'ive', 'hed', 'shed', 'weve', 'theyve']
for w in other_sw:
    stop_words.add(w)

fdist = FreqDist()

for title in tqdm(df.title):
    string = re.sub(r'[^A-Za-z\s]', '', title)
    toks = string.split(' ')
    for tok in toks:
        if tok.lower() not in stop_words and len(tok)>2:
            fdist[tok.lower()] += 1
            
fdist.most_common(20)
```

```python
# lets hope genres will be enough :)
# sparsity will be huge
tokens = []
for word, fr in fdist.items():
    if fr>4:
        tokens.append(word)
len(tokens)
sorted(tokens)[:20]
sorted(tokens)[-20:]
```

```python
vec_df = df.copy()
```

```python
vec_df['imdbUrl'] = vec_df['tconst'].apply(lambda x: 'https://www.imdb.com/title/'+x)
```

```python
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity

mapping = pd.Series(vec_df.reset_index().index, index = vec_df['title'])
X = vec_df.genres.str.split(',', expand=True)[0]
X = pd.get_dummies(X)
X_reg = pd.get_dummies(vec_df.region)
X = pd.concat([X, X_reg], axis=1)
X = pd.concat([X, vec_df.startYear/2020], axis=1)
X = pd.concat([X, vec_df.averageRating/10], axis=1)
X = pd.concat([X, vec_df.isAdult], axis=1)
X = pd.concat([X, vec_df.runtimeMinutes/vec_df.runtimeMinutes.max()], axis=1)
similarity_matrix = cosine_similarity(X, X) #linear_kernel(X, X)
```

```python
type(similarity_matrix)
```

```python
mapping
```

```python
def get_rec(title):
    movie_index = mapping[title]
    similarity_score = list(enumerate(similarity_matrix[movie_index]))
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    similarity_score = similarity_score[1:10]
    #print(similarity_score)
    movie_indices = [i[0] for i in similarity_score]
    return (vec_df.iloc[movie_indices])

get_rec('Crypto')
```

```python
import pickle
with open('similarity_matrix.pkl','wb') as f:
    pickle.dump(similarity_matrix, f)
```
